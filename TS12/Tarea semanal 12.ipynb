#### Anlisis numrico

```python
from pytc2.sintesis_dipolo import remover_polo_dc, remover_polo_infinito
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
from pytc2.remociones import remover_polo_jw, remover_valor_en_infinito, remover_valor_en_dc, remover_polo_sigma
import math
import sympy as sp
import numpy as np
```

```python
s = sp.symbols('s', complex = True)

Z11 = (s**3 + 2*s)/(2*s**2 + 1)
Z11 = sp.nsimplify(Z11)

print_latex(a_equal_b_latex_s('Z11(s)', Z11))
```

```python
# Remocion parcial en cero para llevar el cero a 1/4

Y2, YL1 = remover_polo_dc(1/Z11, omega_zero= sp.Rational(1/4))

print_latex(a_equal_b_latex_s('ZL1', 1/YL1))
print_latex(a_equal_b_latex_s('Y2', Y2))
```

```python
# Removemos el polo finito en 1/4

Z3, Zt, Lt, Ct = remover_polo_jw(1/Y2, omega= sp.Rational(1/4), isImpedance= True)


print_latex(a_equal_b_latex_s('Zt', Zt))
print_latex(a_equal_b_latex_s('Lt', Lt))
print_latex(a_equal_b_latex_s('Ct', Ct))
print_latex(a_equal_b_latex_s('Z3', Z3))
```

Podemos ver que luego de la ltima remocin nos queda una funcin que sabemos que, por su expresin, sabemos que es un inductor. Por condicin de medicin, invertimos y removemos el inductor quedando este en derivacin.

```python
s = sp.symbols('s', complex = True)

Z22 = (2*s**2 +1)/(s*(s**2 + 2))
Z22 = sp.nsimplify(Z22)

print_latex(a_equal_b_latex_s('Z22(s)', Z22))
```

```python
# Remocion parcial en infinito para llevar el cero a 3

Y2, YC1 = remover_polo_infinito(1/Z22, omega_zero= 3)

print_latex(a_equal_b_latex_s('ZC1', 1/YC1))
print_latex(a_equal_b_latex_s('Y2', Y2))
```

```python
# Invertimos y retiramos el polo en 3
Z3, Zt, Lt, Ct = remover_polo_jw(1/Y2, omega= 3, isImpedance= True)

print_latex(a_equal_b_latex_s('Zt', Zt))
print_latex(a_equal_b_latex_s('Lt', Lt))
print_latex(a_equal_b_latex_s('Ct', Ct))
print_latex(a_equal_b_latex_s('Z3', Z3))
```

Por ltimo encontramos $Z3$ que es la impedancia de un capacitor, tal como vimos en el papel, esta impedancia la debemos retirar para que quede en derivacin, por lo que este paso seria invertir y retirar para que quede como mencionamos.
